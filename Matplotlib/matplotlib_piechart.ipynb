1. Write a Python programming to create a pie chart of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
'''

@Author:Naveen Madev Naik
@Date: 2024-08-09
@Last Modified by: Naveen Madev Naik
@Last Modified time: 2024-08-09
@Title :Python programming to create a pie chart of the popularity of programming Languages.

'''


def pie_chart(languages,popularity):

    """
    Description:
        Function will plot the piechart of popularity of programing language

    Parameter:
        languages:list of programing languages
        popularity:popularity of the each programing language

    Return:
        None

    """

    plt.pie(popularity,labels=languages,autopct='%1.1f%%')
    plt.title('Programming Language Popularity')
    plt.axis('equal')
    plt.show()


def main():
    try:
        languages=['Java','Python','PHP','JavaScript','C#','C++']
        popularity=[22.2,17.6,8.8,8,7.7,6.7]
        pie_chart(languages,popularity)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd
```

1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
'''

@Author:Naveen Madev Naik
@Date: 2024-08-13
@Last Modified by: Naveen Madev Naik
@Last Modified time: 2024-08-13
@Title :program to draw a scatter plot for random 1000 x and y coordinates

'''


def scatter_plot(xpoints,ypoints):

    """
    Description:
        Function will draw a scatter plot for random 1000 x and y cooordinates

    Parameter:
        xpoints:x coordinates
        ypoints:y coordinates
      
    Return:
        None    
    """

    fig = go.Figure(data=go.Scatter(x=xpoints, y=ypoints, mode='markers'))
    fig.update_layout(
        title="Scatter Plot of Random Coordinates",
        xaxis_title="X Coordinate",
        yaxis_title="Y Coordinate"
    )
    fig.show()


def main():
    try:
        xpoints = np.random.rand(1000)
        ypoints = np.random.rand(1000)

        scatter_plot(xpoints,ypoints)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
'''

@Author:Naveen Madev Naik
@Date: 2024-08-13
@Last Modified by: Naveen Madev Naik
@Last Modified time: 2024-08-13
@Title :program to draw a scatter plot for random 500 x and y coordinates and style it

'''


def scatter_plot(xpoints,ypoints):

    """
    Description:
        Function will draw a scatter plot for random 500 x and y cooordinates and will add style

    Parameter:
        xpoints:x coordinates
        ypoints:y coordinates
      
    Return:
        None    
    """

    fig = go.Figure(data=go.Scatter(
            x=xpoints,
            y=ypoints,
            mode='markers',
            marker=dict(
                size=10,
                color='rgba(255, 0, 0, 0.6)',  # Semi-transparent red
                symbol='circle',  # Shape of markers
                line=dict(
                    width=2,
                    color='black'  # Border color of markers
                )
            )
        ))

    fig.update_layout(
        title='Styled Scatter Plot of 500 Random Coordinates',
        xaxis_title='X Coordinate',
        yaxis_title='Y Coordinate',
        plot_bgcolor='white',
        xaxis=dict(
            showgrid=True,
            gridcolor='lightgrey'
        ),
        yaxis=dict(
            showgrid=True,
            gridcolor='lightgrey'
        )
    )
    
    fig.show()


def main():
    try:
        xpoints = np.random.rand(500)
        ypoints = np.random.rand(500)

        scatter_plot(xpoints,ypoints)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

4. Write a program to draw a scatter plot for a given dataset and show datalabels on hover
https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

```python
'''

@Author:Naveen Madev Naik
@Date: 2024-08-13
@Last Modified by: Naveen Madev Naik
@Last Modified time: 2024-08-13
@Title :program to draw a scatter plot for a given dataset and show datalabels on hover
        https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

'''


def scatter_plot(xpoints,ypoints):

    """
    Description:
        Function will draw a scatter plot for given dataset and show datalabels on hover

    Parameter:
        xpoints:x coordinates
        ypoints:y coordinates
      
    Return:
        None    
    """

def scatter_plot_with_hover_labels(url):
    """
    Description:
        Function to draw a scatter plot from a dataset with hover labels showing data information.

    Parameter:
        url: URL of the dataset
      
    Return:
        None    
    """
    
    data = pd.read_csv(url)

    fig = go.Figure(data=go.Scatter(
        x=data['Rank'],
        y=data['Population'],
        mode='markers',
        text=data['State'],  
        hoverinfo='text+x+y', 
        marker=dict(
            size=10,
            color='rgba(0, 0, 255, 0.6)',  
            line=dict(
                width=2,
                color='black'
            )
        )
    ))

    fig.update_layout(
        title='Scatter Plot of US States by Population and Area',
        xaxis_title='State Rank',
        yaxis_title='Population',
        plot_bgcolor='white',
        xaxis=dict(
            showgrid=True,
            gridcolor='lightgrey'
        ),
        yaxis=dict(
            showgrid=True,
            gridcolor='lightgrey'
        )
    )

    fig.show()


def main():
    url = 'https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv'
    scatter_plot_with_hover_labels(url)   

if __name__=="__main__":
    main()
```

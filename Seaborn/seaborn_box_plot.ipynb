```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### 1. Write a program to draw box plot of life expectancy by continent for a data set given in a url
https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv<br>

```python
def seaborn_box_plot(url):

    """
    Description:
        Function will draw box plot of life expectancy by continent for given dataset

    Parameter:
        url:url of dataset
      
    Return:
        None    
    """

    data = pd.read_csv(url)

    sns.boxplot(x='continent', y='lifeExp', data=data)

    plt.xlabel('Continent')
    plt.ylabel('Life Expectancy')
    plt.title('Box Plot of Life Expectancy by Continent')

    plt.show()


def main():
    try:
        
        url='https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'
        data=pd.read_csv(url)

        seaborn_box_plot(url)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

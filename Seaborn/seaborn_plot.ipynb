```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

1. Write a program to draw bar plot of sex against survived for a dataset given in the url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
'''

@Author:Naveen Madev Naik
@Date: 2024-08-13
@Last Modified by: Naveen Madev Naik
@Last Modified time: 2024-08-13
@Title :program to draw bar plot of sex against survived for a dataset given in the url
        https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

'''


def seaborn_bar_plot(url):

    """
    Description:
        Function will draw the bar plot of sex against survived for given dataset

    Parameter:
        url:url of dataset
      
    Return:
        None    
    """

    data=pd.read_csv(url) 

    sns.countplot(x='sex', hue='survived', data=data)
    plt.xlabel('Sex')
    plt.ylabel('Survived')
    plt.title('Survival Rate by Sex on the Titanic')

    plt.show()


def main():
    try:
        
        url='https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv'

        seaborn_bar_plot(url)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

2.Write a program to draw a point plot for sex against survived for a dataset given in url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
'''

@Author:Naveen Madev Naik
@Date: 2024-08-13
@Last Modified by: Naveen Madev Naik
@Last Modified time: 2024-08-13
@Title :program to draw point plot for sex against survived for a dataset given in url
        https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

'''


def seaborn_point_plot(url):

    """
    Description:
        Function will draw point plot of sex against survived for given dataset

    Parameter:
        url:url of dataset
      
    Return:
        None    
    """

    data=pd.read_csv(url) 

    sns.pointplot(x='sex', y='survived', data=data)
    plt.xlabel('Sex')
    plt.ylabel('Survived')
    plt.title('Survival Rate by Sex on the Titanic')

    plt.show()


def main():
    try:
        
        url='https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv'

        seaborn_point_plot(url)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

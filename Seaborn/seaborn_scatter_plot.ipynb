```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### 1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv<br>

```python
def seaborn_scatter_plot(url):

    """
    Description:
        Function will draw scatter plot of day against total bill for given dataset

    Parameter:
        url:url of dataset
      
    Return:
        None    
    """
    data=pd.read_csv(url)
    sns.scatterplot(x='day', y='total_bill', data=data)

    plt.xlabel('Day')
    plt.ylabel('Total Bill')
    plt.title('Scatter Plot of Total Bill vs Day')

    plt.show()


def main():
    try:
        
        url='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'

        seaborn_scatter_plot(url)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

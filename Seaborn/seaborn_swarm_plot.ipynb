```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

1. Write a program to draw a swarm plot of total bill against size for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
'''

@Author:Naveen Madev Naik
@Date: 2024-08-13
@Last Modified by: Naveen Madev Naik
@Last Modified time: 2024-08-13
@Title :program to draw a swarm plot of total bill against size for a given dataset
        https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

'''


def seaborn_swarm_plot(url):

    """
    Description:
        Function will draw sworm plot of total bill against size for given dataset

    Parameter:
        url:url of dataset
      
    Return:
        None    
    """

    data = pd.read_csv(url)

    sns.swarmplot(x='size', y='total_bill', data=data)

    plt.xlabel('Size')
    plt.ylabel('Total Bill')
    plt.title('Sworm Plot of Total Bill vs Size')

    plt.show()


def main():
    try:
        
        url='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
        data=pd.read_csv(url)

        seaborn_swarm_plot(url)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

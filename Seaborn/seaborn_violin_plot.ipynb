```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### 1. Write a program to draw a violin plot of sex against total_bill for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv<br>

```python
def seaborn_violin_plot(url):

    """
    Description:
        Function will draw violin plot of day against total bill for given dataset

    Parameter:
        url:url of dataset
      
    Return:
        None    
    """
    data=pd.read_csv(url)
    sns.violinplot(x='day', y='total_bill', data=data)

    plt.xlabel('Day')
    plt.ylabel('Total Bill')
    plt.title('Violin Plot of Total Bill vs Day')

    plt.show()


def main():
    try:
        
        url='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'

        seaborn_violin_plot(url)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

### 2. Write a program to draw a violin plot of species against sepal length for a dataset given in a url
https://github.com/mwaskom/seaborn-data/blob/master/iris.csv<br>

```python
def seaborn_violin_iris_plot(url):

    """
    Description:
        Function will draw violin plot of species against sepal length for given dataset

    Parameter:
        url:url of dataset
      
    Return:
        None    
    """

    df=pd.read_csv(url)
    sns.violinplot(x='species', y='sepal_length', data=df)

    plt.xlabel('Species')
    plt.ylabel('Sepal Length')
    plt.title('Violin Plot of Sepal Length vs Species')

    plt.show()


def main():
    
        
    url= "https://github.com/mwaskom/seaborn-data/raw/master/iris.csv"

    seaborn_violin_iris_plot(url)
 

if __name__=="__main__":
    main()
```
